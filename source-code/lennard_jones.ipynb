```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def lennard_jones(r, ϵ, σ):
    return 4.0*ϵ*((σ/r)**12 - (σ/r)**6)
```

```python
r = np.linspace(0.001, 0.2, 101)
V = lennard_jones(r, 0.1, 1.0e-1/2.0**(1/6))
```

```python
plt.plot(r, V)
plt.ylim(-0.2, 1.0)
plt.xlabel('$r$')
plt.ylabel('$V$');
```

```python
particles = np.random.uniform(size=(1_000, 3))
```

```python
def compute_distances(particles):
    n = particles.shape[0]
    distances = []
    for i in range(n):
        for j in range(i + 1, n):
            distances.append(np.sqrt(np.sum((particles[i] - particles[j])**2)))
    return np.sort(distances)
```

```python
distances = compute_distances(particles)
```
